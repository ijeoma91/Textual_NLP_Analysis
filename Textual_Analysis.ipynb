 #### **Step 5.2 Text Tokenization**.
 
 
 After transforming the customer complainst text into lower case, we will now tokenize it using NLTK's Regexptoken to perform our tokenization

```python

from nltk.tokenize import RegexpTokenizer

regexp = RegexpTokenizer('\w+')
Text_df["Customer_Text_Token"] = Text_df['Customer_Complaints_Text']. apply(regexp.tokenize)
Text_df.head(6)
```

```python
Text_df['Customer_Complaints_Text'] = Text_df['Customer_Complaints_Text'].astype(str).str.lower()
Text_df.head(5)
```

```python
Text_df['Customer_Text_Token'] = Text_df['Customer_Text_Token'].apply(lambda x: [words for words in x if words not in stopwords])
Text_df.head(6)
```

```python
Text_df["Clean_text_string"] = Text_df["Customer_Text_Token"].apply(lambda x: ' '.join([item for item in x if len(item)>6]))
Text_df[["Customer_Complaints_Text", "Customer_Text_Token", "Clean_text_string"]].head()
```
