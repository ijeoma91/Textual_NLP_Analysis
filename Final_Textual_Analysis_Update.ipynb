### **Stemming:**

**A.** **Using the Stemming Techniques:** Stemming algorithm is a techniques used to reduces a word. For example,  reference to our output  above we can see that the stemming techniques reduced words like:
original word = (negative)	Stem= ((neg), 	(negat ))




```python
from nltk.stem import PorterStemmer
#from nltk.corpus import stopwords
import re
# First we will tokenie each word from the clean_text_string in row 2
words  = nltk.tokenize.WhitespaceTokenizer().tokenize (Text_df["Clean_text_string"].to_string())
# New dataframe is created to store those word called Original_Text_Word
DataFrame_ = pd.DataFrame()
DataFrame_['Original_Text_word'] = pd.Series(words)
# We will use porter's stemmer and add new column. Porter stemming is suffix stripping algorithm. In the column of Porter Stemming every word will respresent  a sequence of consonants/vowels.
PorterStemmedWords = [nltk.stem.PorterStemmer().stem(word) for word in words]
DataFrame_['PorterStemmedWords'] = pd.Series(PorterStemmedWords)
#We will also use SnowBall stemmer to process or text which is a better version of the porter stemmer.
# viewing the output we can see that SnowballStemmer words is much better than PoterStemmer.

SnowBallStemmedWords = [nltk.stem.SnowballStemmer("english").stem(word) for word in words]
DataFrame_['SnowballStemmedWords'] = pd.Series(SnowBallStemmedWords)
DataFrame_.head(10)
```

### **5.4 Lemmatization**

Here we will implement Lemmatization technique. Our arm is to remove inflectional endings and return Lemma. The reason why this technique is implemented is because, Lemmatization is more accurate than Stemming. The two techniques will be implemented and the output will be compared  to see the difference.

```python
import nltk
nltk.download('wordnet')

#LEMMATIZATION with WordNet (We will use our target text row 2  in column Clean_Text_string)
Row_2words  = nltk.tokenize.WhitespaceTokenizer().tokenize(Text_df["Clean_text_string"].to_string())
dframe = pd.DataFrame()
dframe['OriginalWords'] = pd.Series(Row_2words)
#We will use WordNet Lemmatization to lemmatize word bease on its context and usage.
wordNetLemmatizedWords = [nltk.stem.WordNetLemmatizer().lemmatize(word) for word in Row_2words]
dframe['WordNetLemmatizer'] = pd.Series(wordNetLemmatizedWords)
dframe.head(10)
```

## **TF-IDF (Term Frequency Inverse Document Frequency)**

This techniques is an extension of BoW method which provides a numerical statistic  that reflect no the importance of a specific word which is related to the whole dataset collection.. Frequently used word are not necessary important in the dataset. Implementing TF-IDF we can relates a specific word to a specific document in the whole corpus and thereby assigning the importance to those word.The higher the values of the TF-IDF, the more significant rare those words are to the document.

### **Topic Modelling**

This type of technique will help in finding similarities in our dataset thereby discovering abstract topic. This method can be used to filter words.

In this phase we will use two methods and then compare the both results

1. **Latent Semantic Algorthm Analysis:** This is unsupervised techniques used in NLP information intrieval and this techniques can also be used to analyze the relationship between a set of word



```python
from sklearn.decomposition import TruncatedSVD # Using Sklean to implement our LSA algorithm


vectorizer = TfidfVectorizer(use_idf=True, max_features= 15, smooth_idf=True) # Initiating our Tfid algorithm using sklearn
model_df = vectorizer.fit_transform([Text_df["Clean_text_string"].to_string()]) # Fitting in our data

LSA_model = TruncatedSVD(n_components=15, algorithm='randomized', n_iter = 10) # Setting our paramater example like how many features I want to extract

lsa = LSA_model.fit_transform(model_df) # Fitting our data for interation

l =lsa[0]
print('Analysis 0 :')
for i, topic in enumerate(l):# Data iteration
  print("Topic", "i" , ":", topic*100) # Print out every interation

data_2 = pd.DataFrame(model_df.toarray(), columns=vectorizer.get_feature_names_out()) # Creating a datafrme for a better reading

data_2
```

1. **Latent Dirichlet Allocation(LDA):**


This method is used to generate topic based on word frequency of documents.






```python
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import LatentDirichletAllocation


vectorizer = TfidfVectorizer(use_idf=True,
max_features=15, # Here we are only extracting 15 features
smooth_idf=True)
model_1 = vectorizer.fit_transform([Text_df["Clean_text_string"].to_string()])

lda_model=LatentDirichletAllocation(n_components=2,learning_method='online',random_state= 100) # Fitting in our dataset (Column)

lda_top=lda_model.fit_transform(model_1) # Fitting in our data
for x,topic in enumerate(lda_top[0]):# Iteration
  print("Topic ",x,": ",topic*100,"%")

data_3 = pd.DataFrame(model_1.toarray(), columns=vectorizer.get_feature_names_out()) # Creating a dataframe for a better and clear output reading

data_3
```

### **Output Interpretation:**
15 features that was extracted, this shows the weights based on the topics. this implies which of the words dominate the topics
